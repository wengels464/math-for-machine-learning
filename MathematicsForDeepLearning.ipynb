## 2.1 - Linear Algebra

Algebra itself is defined as the construction of objects (symbols) and operations to manipulate them (algebraic laws).

Linear algebra is the study of vectors and the rules pertaining to the manipulation of vectors.

All polynomials are vectors, but we are primarily interested in geometric (straight) vectors.

Audio signals are also vectors, as they are expressed as a series of numbers.

In the below case the column vector a is defined as being in the 3-dimensional vector space.

$a = \begin{bmatrix}1 \\ 2 \\ 3 \end{bmatrix} \in \mathbb{R}^3$

More generally:

$v_n = \begin{bmatrix}x_0 \\ ... \\ x_{n-1} \\ x_{n} \end{bmatrix} \in \mathbb{R}^n$

For vector addition, two or more vectors added in the same vector space will yield a single new vector in the same vector space.

$a,b \in \mathbb{R}^n$

$a + b = c \in \mathbb{R}^n$

**Scalar multiplication works comparably**

For $\lambda \in \mathbb{R}$ and $a \in \mathbb{R}^n$ the result is scaled vector $\lambda a \in \mathbb{R}^n$.

For our purposes, we will focus only on finite-dimensional vector spaces, in which there is a 1:1 correspondence between any kind of vector and $\mathbb{R}^n$.

## 2.2 - Matrices

We can define a generic matrix formally:
$\mathbf{A} = \begin{bmatrix} a_11 & a_12 & ... & a1n \\
                              a_21 & a_22 & ... & a2n \\
                              ... & ... & ... & ...\\
                              a_m1 & a_m2 & ... & a_mn \end{bmatrix},\quad a_ij \in \mathbb{R}$

A matrix is typically defined as having $m$ rows and $n$ columns where $m,n \in \mathbb{N}$.

Matrices with only one dimension are columned **column** or **row** matrices.

$\mathbb{R}^{m \times n}$ represents the set of all real-valued $m,n$ matrices.

For **matrix addition** we simply add each element in $\mathbf{A}_{m \times n}$ and $\mathbf{B}_{m \times n}$ to yield:
$\mathbf{C}_{m \times n}$.

**Matrix multiplication** is more complex, requiring matrices of compatible dimensions. 

In this case, if we have two matrices to multiply $\mathbf{A}_{n \times k}$ and $\mathbf{B}_{k \times m}$ this yields a new matrix: $\mathbf{C}_{n \times m}$.

*Notice that the outer dimensions of the two matrices yield the new dimensions of C matrix.*

*Notice that the inner dimensions of the two matrices **determine** whether or not the matrices can be multiplied*

It is important to note that unlike constants and variables, matrix multiplication is **not commutative**.

In other words $\mathbf{A}\mathbf{B}\neq\mathbf{B}\mathbf{A}$.

We must also define an **identity matrix**. 

These are matrices that have all zeros except for a diagonal of 1s stretching from top left to bottom right.

Matrices in algebra enjoy the following properties:

**Associativity**: $(\mathbf{A} \mathbf{B})\mathbf{C} =  \mathbf{A}(\mathbf{B} \mathbf{C})$

**Distributivity**: $(\mathbf{A} + \mathbf{B})\mathbf{C} = \mathbf{A}\mathbf{C} + \mathbf{B}\mathbf{C}$

**Identity**: $\mathbf{I}_n \mathbf{A} = \mathbf{A} \mathbf{I}_n = \mathbf{A}$
